Из этой таблицы можем сделать несколько выводов:
1. всего 891 пассажиров - 577 мужчин и 314 женщин
2. 38 % выжило
3. средний возраст - 30 лет
4. самому старому человеку 80 лет, а младшему еще нет года
5. пассажиры жили в 147 разных номерах
6. большинство(644 человека) сели на борт в порту S (Southampton)
7. средняя стоимость билета 32$ 
8. самый дешевый билет оказался бесплатным, а самый дорогой стоит 512$

```python
survived = df[df["survived"] == 1]

all_survived = survived.shape[0]
mask = (survived['adult_male'])
old_men = survived[mask].shape[0]

print(f"survived old men: {old_men}, all survived passengers: {all_survived}")
print("persent of survived old men:", round(old_men / all_survived * 100, 3))

survived.head()
```

процент взрослых мужчин среди выживших уменьшился в 2.5 раза


```python
survived.describe(include='all')
```

из этих данных можно узнать что:
1. выжило всего 342 человека - 233 женщины(68%) и 109 мужчин(38%), то есть осталось в живых (233/314 * 100) 74% женщин и только (109/577 * 100) 19% мужчин
2. средний возраст почти не отличается от исходного - 28 лет
3. средняя стоимость билета выросла на 16$, а это значит, что больше шансов выжить с дорогим билетом
