```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function_2d(x, y):
    return 5*x**3 + 4*x* y - 2*y**2 + 1

def gradient_2d(x, y):
    return np.array((15*x**2 + 4 * y, 4*x - 4*y))

def step(point, lr):
    return point - lr * gradient_2d(*point)

def gradient_descent_alg(x_0, y_0, lr, num_steps):
    history = [np.array((x_0, y_0))]
    gradient_history = [] 
    for i in range(1, num_steps + 1):
        gradient_history.append(gradient_2d(*history[-1]))
        history.append(step(np.array(history[i - 1]), lr))
        
    return (np.array(history), np.array(gradient_history))
```

```python
x0, y0 = 1, -2

g1 = gradient_2d(x0, y0)
print(g1)
x1, y1 = step((x0, y0), 0.1)
print(x1, y1)
print()

g2 = gradient_2d(x1, y1)
print(g2)
x2, y2 = step((x1, y1), 0.1)
print(x2, y2)
```

```python
def gradient(x):
    return 8*x**7-20*x**3+8*x-1

def step(x, lr):
    return x - lr * gradient(x)

def gradient_descent_alg(x_0, lr, num_steps):
    history = [x_0]
    gradient_history = [] 
    for i in range(1, num_steps + 1):
        history.append(step(history[i - 1], lr))
        
    return np.array(history)[-1]
```

```python
x_answers = []
for start in np.arange(-1.5, 1.5, 0.25):
    x_answers.append(gradient_descent_alg(start, 0.01, 100))

x_answers
```

```python
plt.plot(X, y)

y_ans = list(map(function, x_answers))
plt.scatter(x_answers, y_ans, c='r')
```
